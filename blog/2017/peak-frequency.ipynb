```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
np.random.random(size=2000)/10
```

```python
def create_sine(frequency):
    return np.sin(frequency*np.linspace(0, 2*np.pi, 500)) + np.random.random(size=500)/20
```

```python
plt.plot(create_sine(30))
#plt.plot(np.sin(np.linspace(0, 400*np.pi, 2000)))
```

```python
train_y = np.tile(np.arange(1, 100, 1), reps=200)
```

```python
train_x = np.array([create_sine(x) for x in train_y])
```

```python
train_x.shape
```

```python
from keras.models import Model
from keras.layers import Dense, Input, LSTM
```

```python
input_series = Input(shape=(500,),name='Input')
dense_1 = Dense(100)(input_series)
pred = Dense(1, activation='relu')(dense_1)
model = Model(input_series, pred)
SVG(model_to_dot(model, show_shapes=True).create(prog='dot', format='svg'))
```

```python
model.compile('adam','mean_absolute_error')
```
