```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/tidyverse/ggplot2/master/data-raw/diamonds.csv")
```

```python
from altair import *
Chart(df.head(1000)).mark_line().encode(x='price:Q', y='cut:N')
```

```python
ngroups = len(df.cut.unique())
fig, ax = plt.subplots(nrows=ngroups)
for i, (group_name, group) in enumerate(df.groupby("cut")):
    group['price'].plot(kind='kde', ax=ax[i])
```
